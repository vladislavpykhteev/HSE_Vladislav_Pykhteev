```python
"""
Задание 1 -  Создайте ряд функций для проведения математических вычислений:
1) функция вычисления факториала числа (произведение натуральных чисел от 1 до
n). Принимает в качестве аргумента число, возвращает его факториал;
2) Поиск наибольшего числа из трёх. Принимает в качестве аргумента кортеж из трёх
чисел, возвращает наибольшее из них;
3) расчёт площади прямоугольного треугольника. Принимает в качестве аргумента
размер двух катетов треугольника. Возвращает площадь треугольника
"""

# функция факториала числа
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result

# функция поиска наибольшего из трёх чисел
def findMax(numbers):
    maxNum = numbers[0]
    if numbers[1] > maxNum:
        maxNum = numbers[1]
    if numbers[2] > maxNum:
        maxNum = numbers[2]
    return maxNum

# функция расчёта площади прямоугольного треугольника
def triangleArea(leg1, leg2):
    area = (leg1 * leg2) / 2
    return area

# проверяем как работают функции
print("Факториал 5 =", factorial(5))
print("Факториал 7 =", factorial(7))
print()

myNumbers = (10, 25, 7)
print("Числа:", myNumbers)
print("Наибольшее:", findMax(myNumbers))
print()

print("Площадь треугольника с катетами 6 и 8 =", triangleArea(6, 8))
```
