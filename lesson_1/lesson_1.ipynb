```python
"""
Задание 1 - работа с переменными
Создаю переменные разных типов и вывожу на экран
Потом запрашиваю данные у пользователя
Проверяю функцию id() - как меняется адрес объекта в памяти
"""

# Задание 1 - работа с переменными
helloText = "Привет! Это моя первая разработка на Python!"
age = 26
versionDO = 2.1
boolType = True


print("Созданные переменные:")
print(helloText)
print("Мне ", age, " лет")
print("Я имею опыт работы с 1С Документооборот версии ", versionDO)
print("Я хочу научиться работать с Python, это - ", boolType)
print()

# запрашиваем у пользователя данные
userName = input("Введите ваше имя: ")
userCity = input("Введите ваш город: ")
userNumber = input("Введите любое число: ")

print("Вы ввели следующее:")
print("Имя - ", userName)
print("Город - ", userCity)
print("Число - ", userNumber)
print()

# наблюдаем за id() - неизменяемые типы (int, str) меняют адрес при изменении
# изменяемые типы (list) сохраняют адрес
myInt = 10
print("Число до изменения - ", myInt, "  id =", id(myInt))
myInt = myInt + 5
print("Число после изменения - ", myInt, "  id =", id(myInt))
print("Видно что id изменился - создался новый объект в памяти!")
print()

myStr = "Привет"
print("Строка до изменения - ", myStr, "  id =", id(myStr))
myStr = myStr + " Python"
print("Строка после изменения - ", myStr, "  id =", id(myStr))
print("Строки тоже неизменяемые - id поменялся")
print()

myList = [1, 2, 3]
print("Список до изменения - ", myList, "  id =", id(myList))
myList.append(4)
print("Список после изменения - ", myList, "  id =", id(myList))
print("Список изменяемый - id остался тем же!")
```

```python
"""
Задание 2 - перевод секунд в часы минуты и секунды
Пользователь вводит количество секунд
Проверяю что введены только цифры через isdigit()
Перевожу строку в число через int()
Считаю часы минуты секунды и вывожу по отдельности
"""

timeInput = input("Введите время в секундах: ")

if timeInput.isdigit():
    totalSeconds = int(timeInput)

    hours = totalSeconds // 3600
    minutes = (totalSeconds % 3600) // 60
    seconds = totalSeconds % 60

    print()
    print("Результат:")
    print("Часы - ", hours)
    print("Минуты - ", minutes)
    print("Секунды - ", seconds)

else:
    print("Ошибка - нужно ввести только цифры!")
```

```python
"""
Задание 3 - считаем сумму n + nn + nnn
Пользователь вводит цифру от 1 до 9
Например если ввел 3 то считаем 3 + 33 + 333 = 369
Умножение строки на число в Python - это повторение строки
"3" * 2 = "33", "3" * 3 = "333"
Потом переводим обратно в int и складываем
"""

nInput = input("Введите цифру от 1 до 9: ")

if nInput.isdigit() and 1 <= int(nInput) <= 9:
    n = int(nInput)
    nn = int(nInput * 2)
    nnn = int(nInput * 3)

    result = n + nn + nnn

    print()
    print("n = ", n)
    print("nn = ", nn)
    print("nnn = ", nnn)
    print("Результат: ", n, "+", nn, "+", nnn, "=", result)

else:
    print("Ошибка - введите одну цифру от 1 до 9!")
```
