```python
"""
Задание 2 - структура данных "Участники спора"
Каждый словарь хранит информацию об отдельном участнике. Словарь содержит пары
«ключ-значение» параметрами, то есть характеристиками участника: наименование,
статус, ИНН.
Данные вводятся пользователем через консоль для трёх участников
Полученные данные выводятся в консоль
"""

# создаем пустой список участников
disputeMembers = []

# вводим данные для трёх участников
print("Введите данные участников спора:")
print()

print("Участник 1:")
name1 = input("Наименование: ")
status1 = input("Статус (Истец / Ответчик / Третье лицо): ")
inn1 = input("ИНН: ")

print()
print("Участник 2:")
name2 = input("Наименование: ")
status2 = input("Статус (Истец / Ответчик / Третье лицо): ")
inn2 = input("ИНН: ")

print()
print("Участник 3:")
name3 = input("Наименование: ")
status3 = input("Статус (Истец / Ответчик / Третье лицо): ")
inn3 = input("ИНН: ")

# формируем словари и добавляем в список
member1 = {"name": name1, "status": status1, "inn": inn1}
member2 = {"name": name2, "status": status2, "inn": inn2}
member3 = {"name": name3, "status": status3, "inn": inn3}

disputeMembers.append(member1)
disputeMembers.append(member2)
disputeMembers.append(member3)

# выводим результат
print()
print("Список участников спора:")
print(disputeMembers)
```
