```python
# Домашнее задание 4 - функция для валидации ИНН
# Функция принимает в качестве аргумента строку, содержащую ИНН
# или просто набор цифр, похожий на ИНН и проверяет правильность ИНН
# математически, без обращения к ЕГРЮЛ.
# ИНН организации - 10 цифр, ИНН физлица/ИП - 12 цифр

# вспомогательная функция - вычисляет контрольную сумму по коэффициентам
def calcCheckSum(innDigits, coefficients):
    total = 0
    for i in range(len(coefficients)):
        total = total + innDigits[i] * coefficients[i]
    return total

# вспомогательная функция - вычисляет контрольное число из суммы
def calcControlNumber(checkSum):
    controlNum = checkSum % 11
    if controlNum > 9:
        controlNum = controlNum % 10
    return controlNum

# проверка ИНН организации (10 цифр)
def validateInnOrg(innDigits):
    # коэффициенты для 10-значного ИНН
    coefficients10 = [2, 4, 10, 3, 5, 9, 4, 6, 8]

    checkSum = calcCheckSum(innDigits, coefficients10)
    controlNum = calcControlNumber(checkSum)

    # сравниваем контрольное число с 10-й цифрой ИНН (индекс 9)
    if controlNum == innDigits[9]:
        return True
    else:
        return False

# проверка ИНН физлица / ИП (12 цифр)
def validateInnPerson(innDigits):
    # коэффициенты для первого контрольного числа (11-й знак)
    coefficients11 = [7, 2, 4, 10, 3, 5, 9, 4, 6, 8]

    # коэффициенты для второго контрольного числа (12-й знак)
    coefficients12 = [3, 7, 2, 4, 10, 3, 5, 9, 4, 6, 8]

    checkSum1 = calcCheckSum(innDigits, coefficients11)
    controlNum1 = calcControlNumber(checkSum1)

    checkSum2 = calcCheckSum(innDigits, coefficients12)
    controlNum2 = calcControlNumber(checkSum2)

    # проверяем оба контрольных числа
    if controlNum1 == innDigits[10] and controlNum2 == innDigits[11]:
        return True
    else:
        return False

# основная функция валидации - принимает строку с ИНН
def validateInn(innStr):
    # убираем пробелы на всякий случай
    innStr = innStr.strip()

    # проверяем что строка содержит только цифры
    if not innStr.isdigit():
        print("Ошибка - ИНН должен содержать только цифры")
        return False

    # проверяем длину - должна быть 10 или 12 цифр
    if len(innStr) != 10 and len(innStr) != 12:
        print("Ошибка - ИНН должен содержать 10 или 12 цифр, передано:", len(innStr))
        return False

    # переводим строку в список цифр
    innDigits = []
    for ch in innStr:
        innDigits.append(int(ch))

    # вызываем нужную проверку в зависимости от длины
    if len(innStr) == 10:
        return validateInnOrg(innDigits)
    else:
        return validateInnPerson(innDigits)

# тестируем функцию
print("Тестирование валидации ИНН:")
print()

# валидные ИНН организаций
testInn1 = "7732005842"
testInn2 = "6671255706"
testInn3 = "7720300064"

# валидные ИНН физлиц / ИП (12 цифр)
testInn4 = "667400000000"   # пример ИНН ИП
testInn5 = "500100732259"   # известный валидный 12-значный ИНН

# невалидные ИНН
testInn6 = "1234567890"     # неправильное контрольное число
testInn7 = "123"            # слишком короткий
testInn8 = "7732005843"     # последняя цифра изменена - должно быть False

print("ИНН", testInn1, "- результат:", validateInn(testInn1))
print("ИНН", testInn2, "- результат:", validateInn(testInn2))
print("ИНН", testInn3, "- результат:", validateInn(testInn3))
print("ИНН", testInn4, "- результат:", validateInn(testInn4))
print("ИНН", testInn5, "- результат:", validateInn(testInn5))
print("ИНН", testInn6, "- результат:", validateInn(testInn6))
print("ИНН", testInn7, "- результат:", validateInn(testInn7))
print("ИНН", testInn8, "- результат:", validateInn(testInn8))
```
